## 1. Series

Pandas의 Series는 1차원 배열로서 다음의 특징을 가진다.

## 2. 인덱싱 (indexing)

Series 생성시 0부터 순차적으로 부여되는 index를 확인할 수 있다.

이를 `RangeIndex`라 부른다.

> Start: 첫 번째 인덱스 값
>
> Stop: 데이터의 길이 (len(data))
>
> Step: 인덱스 간의 증가 값 (기본값 1)

이때, index의 값으로 Series의 특정 위치의 요소를 불러올 수 있다.

````python
s = pd.Series(['부장', '차장', '대리', '사원', '인턴'])
s
s.index
````

````python
s[2]
````

### fancy indexing

**팬시 인덱싱(Fancy Indexing)**은 정수 리스트나 배열을 사용하여 여러 개의 특정 행이나 열을 한 번에 선택하는 방법이다.

일반적인 단일 인덱싱과 달리, 리스트나 배열을 사용하여 특정 위치의 요소를 선택할 수 있다.

````python
s[[1, 3]]
````

````python
s[np.arange(1, 4, 2)]
````

### boolean indexing

조건식을 만들어서 특정 조건에 대하여 `True`에 해당하는 값만 필터링 할 수 있다.

s를 정의 후

````python
np.random.seed(0)
s = pd.Series(np.random.randint(10000, 20000, size=(10,)))
s
````

다음과 같이 True, False로 이루어진 Boolean series를 생성 후 index로 활용하여 필터할 수 있다.

필터는 다음과 같이 사용하면 True인 값을 가져올 수 있다.

````python
s > 15000
````

````python
s[s > 15000]
````

기본 값으로 부여되는 `RangeIndex`에 사용자 정의의 `index`를 **지정**할 수 있다.

````python
s = pd.Series(['마케팅', '경영', '개발', '기획', '인사'], index=['a', 'b', 'c', 'd', 'e'])
s
````

````python
s.index
````

사용자 정의의 index 부여시 **변경된 index**로 조회 가능하다.

````python
s['c']
````

````python
s[['a', 'd']]
````

정의할 때 지정하는 방법 외에,
먼저 `Series`를 생성 후 index 속성 값에 새로운 index를 할당하여 인덱스를 지정할 수도 있다.

````python
s = pd.Series(['마케팅', '경영', '개발', '기획', '인사'])
s.index
````

````python
s.index = list('abcde')
s
````

## 3. 속성 (attribute)

### values

`values`는 Series 데이터 값(value)만 **numpy array** 형식으로 가져 온다.

````python
s.values
````

### ndim - 차원
Series는 1차원 자료구조이기 때문에 ndim 출력시 **1**이 출력된다.

````python
s.ndim
````

### shape

shape은 데이터의 모양(shape)을 알아보기 위하여 사용하는데, Series의 shape은 **데이터의 갯수**를 나타낸다.

**튜플(tuple)** 형식으로 출력된다.

````python
s.shape
````

### NaN (Not a Number)

Pandas에서 **NaN 값**은 비어있는 **결측치 데이터**를 의미합니다.

임의로 비어있는 값을 대입하고자 할 때는 **numpy의 nan (np.nan)**을 입력합니다.

````python
s = pd.Series(['선화', '강호', np.nan, '소정', '우영'])
s
````

### 결측치 (NaN) 값 처리

`isnull()`과 `isna()`은 **NaN** 값을 찾는 함수이다.

`isnull()`과 `isna()`는 결과가 동일하다.

````python
s.isnull()
````

````python
s.isna()
````

이를 boolean indexing에 적용해볼 수 있다.

````python
s[s.isnull()]
````

````python
s[s.isna()]
````

`notnull()`은 NaN값이 아닌, 즉 비어있지 않은 데이터를 찾는 함수 입니다.

````python
s.notnull()
````

````python
s.notna()
````

NaN 값을 제외한, 나머지 값을 가져오려면 다음과 같이 할 수 있다.

````python
s[s.notnull()]
````

이제, `sample` 중 결측치 데이터만 필터해보자

````python
sample = pd.Series(['IT서비스', np.nan, '반도체', np.nan, '바이오', '자율주행'])
sample
````

## 5. DataFrame

[도큐먼트](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html)

1차원 데이터 구조 보다는, 주로 많이쓰이는 Excel 데이터 시트와 같은 구조를 가지는 DataFrame에 대해서 알아보자.

`pd.DataFrame`

> 2차원 데이터 구조 (Excel 데이터 시트를 생각하면 된다)
>
> 행(row), 열(column)으로 구성되어 있다.
>
> 각 열(column)은 각각의 데이터 타입 (dtype)을 가진다.

### 생성

**list 를 통해 생성**할 수 있다. DataFrame을 만들 때는 **2차원 list를 대입**한다.

````python
pd.DataFrame([[1, 2, 3],
              [4, 5, 6],
              [7, 8, 9]])
````

아래 예제와 같이 **columns를 지정**하면, DataFrame의 각 열에 대한 컬럼명이 붙는다.

````python
pd.DataFrame([[1, 2, 3],
              [4, 5, 6],
              [7, 8, 9]], columns=['가', '나', '다'])
````

**dictionary를 통한 생성**도 가능하다.

편리한 점은 dictionary의 **key 값이 자동으로 column 명으로 지정**된다는 점이다.

````python
data = {
    'name': ['Kim', 'Lee', 'Park'],
    'age': [24, 27, 34],
    'children': [2, 1, 3]
}
df = pd.DataFrame(data)
df
````

### 속성 (Attribute)

DataFrame은 다음의 **속성**을 가진다.

- **index**: index (기본 값으로 RangeIndex)
- **columns**: column 명
- **values**: numpy array형식의 데이터 값
- **dtypes**: column 별 데이터 타입
- **T**: DataFrame을 전치(Transpose)

````python
df.index
````

````python
df.columns
````

````python
df.values
````

데이터프레임에서의 `dtypes` 속성은 컬럼별 dtype을 출력한다.

````python
df.dtypes
````

````python
df.T
````

### index 지정

데이터프레임에서도 Series와 마찬가지로 index를 지정할 수 있다.

````python
df
````

````python
df.index = list('abc')
df
````

(참고) DataFrame의 indexing / slicing은 나중에 세부적으로 다루도록 하겠다.

### column 다루기

DataFrame에 key 값으로 column의 이름을 지정하여 column을 선택할 수 있다.

1개의 column을 가져올 수 있으며, **1개의 column 선택시 Series**가 된다.

````python
df['name']
````

````python
type(df['name'])
````

2개 이상의 column 선택은 **fancy indexing으로 가능**하다.

````python
df[['name', 'children']]
````

(참고) column에 대한 slicing도 가능 하지만 이 부분도 나중에 다루도록 하겠다.

**rename**으로 column명 변경 가능하다.


DataFrame.rename(columns={'바꾸고자 하는 컬럼명': '바꿀 컬럼명'})

````python
df.rename(columns={'name': '이름'})
````

`inplace=True` 옵션으로 변경사항을 바로 적용할 수 있다.

````python
df.rename(columns={'name': '이름'}, inplace=True)
df
````
