## 1. copy

DataFrame을 다룰때, 원본에 영향을 주어 파일을 다시 불러와야 하는 상황이 발생한다. 이때 copy 함수로 원본에 영향을 주지 않고, DataFrame을 수정할 수 있다.

copy 함수는 DataFrame을 **복제**한다. 복제한 DataFrame을 수정해도 **원본에는 영향을 미치지 않는다.**

````python
df.head()
````

`copy()`로 DataFrame을 복제한다.

````python
df_copy = df.copy()
````

````python
df_copy.head()
````

`df_copy`의 `age`를 99999로 임의 수정해보자.

````python
df_copy.loc[0, 'age'] = 99999
````

df_copy에는 수정사항이 반영된 것을 확인할 수 있다.

````python
df_copy.head()
````

하지만, 원본 DataFrame의 **데이터는 변경되지 않고 그대로 남아** 있다.

````python
df.head()
````

## 2. 결측치

결측치는 **비어있는 데이터**를 의미한다.

결측치에 대한 처리는 매우 중요하다.

결측치에 대한 처리를 해주려면 **다음의 내용**을 반드시 알아야 한다.

1. 결측 데이터 확인
2. 결측치가 **아닌** 데이터 확인
3. 결측 데이터 **채우기**
4. 결측 데이터 **제거하기**

### 결측치 확인 - isnull(), isna()

컬럼(column)별 결측치의 갯수를 확인하기 위해서는 `sum()` 함수를 붙혀주면 된다.

`sum()`은 Pandas의 통계 관련 함수로 개수의 총 합을 알 수 있다.

**isnull()**

````python
df.isnull().sum()
````

**isna()**

isnull() 과 동작이 완전 같다. 편한 것으로 사용하면 된다. (심지어 도큐먼트도 같다)

````python
df.isna().sum()
````

DataFrame 전체 결측 데이터의 갯수를 합산하기 위해서는 `sum()`을 두 번 사용하면 된다.

````python
df.isnull().sum().sum()
````

### 결측치가 아닌 데이터 확인 - notnull()

`notnull()`은 `isnull()`과 정확히 **반대** 개념이다.

결측치가 아닌 데이터를 확인하기 위해서는 `notnull()`을 사용하면 된다.

````python
df.notnull().sum()
````

### 결측 데이터 필터링

`isnull()` 함수가 결측 데이터를 찾는 **boolean index** 이다.

즉, `loc`에 적용하여 조건 필터링을 걸 수 있다.

````python
df.loc[df['age'].isnull()]
````

### 결측치 채우기 - fillna()

`fillna()`를 활용하면 결측치에 대하여 **일괄적으로 값을 채울 수** 있다.

````python
df1.tail()
````

888번 index의 **결측치가 700으로 채워**진 것을 확인할 수 있다.

````python
df1['age'].fillna(700).tail()
````

### NaN 값이 있는 데이터 제거하기 (dropna)

````python
df1 = df.copy()
````

````python
df1.tail()
````

`dropna()`로 **1개 라도 NaN 값이 있는 행**은 제거할 수 있다. (`how='any'`)

````python
df1.dropna()
````

기본 옵션 값은 `how=any`로 설정되어 있으며, 다음과 같이 변경할 수 있다.

> **any**: 1개 라도 NaN값이 존재시 drop
>
> **all**: 모두 NaN값이 존재시 drop

````python
df1.dropna(how='all')
````
