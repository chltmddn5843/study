[도큐먼트](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DatetimeIndex.year.html)

- pandas.Series.dt.year: 연도
- pandas.Series.dt.month: 월
- pandas.Series.dt.day: 일
- pandas.Series.dt.hour: 시
- pandas.Series.dt.minute: 분
- pandas.Series.dt.second: 초
- pandas.Series.dt.microsecond: micro 초
- pandas.Series.dt.nanosecond: nano 초
- pandas.Series.dt.week: 주
- pandas.Series.dt.weekofyear: 연중 몇 째주
- pandas.Series.dt.dayofweek: 요일
- pandas.Series.dt.weekday: 요일 (dayofweek과 동일)
- pandas.Series.dt.dayofyear: 연중 몇 번째 날
- pandas.Series.dt.quarter: 분기

## 7. pd.cut() - 구간 나누기(binning)

연속된 수치(continuous values)를 **구간으로 나누어 카테고리화** 할 때 사용한다.

````python
df2.head()
````

직접 범위 설정을 해줄 수 있다.
- `right=False`로 지정시 우측 범위를 포함하지 않는다.

````python
bins = [0, 6000, 100000, df2['이동거리'].max()]
pd.cut(df2['이동거리'], bins, right=False)
````

`labels`를 지정해 줄 수 있으며, 지정한 bins의 개수보다 1 개가 적어야 한다.

````python
labels = ['적음', '보통', '많음']
````

````python
pd.cut(df2['이동거리'], bins, labels=labels, right=False)
````

`pd.cut()`을 활용하여 쉽게 그룹을 나눌 수 있다.

````python
df2.head()
````

`bins` 옵션에 나누고자 하는 **구간의 개수**를 설정한다.

````python
df2['이동거리_cut'] = pd.cut(df2['이동거리'], bins=3)
````

````python
df2['이동거리_cut'].value_counts()
````

분포를 보니 첫 구간에 대부분의 데이터가 쏠려 있다. 딱봐도 올바르지 않은 방법 같아 보인다.

`pd.cut()`은 **최소에서 최대 구간을 지정한 bin만큼 동일하게 분할** 하기 때문에 이런 현상이 발생할 수 있다.

고르게 분포한 데이터라면 괜찮지만, 튀는 **이상치(outlier)가 있는 경우에는 안 좋은 결과**를 초래 한다.

## 8. pd.qcut() - 동일한 갯수를 갖도록 구간 분할

`pd.cut()`과 유사하지만, **quantity 즉 데이터의 분포를 최대한 비슷하게 유지**하는 구간을 분할 한다.

````python
df2['이동거리_qcut'] = pd.qcut(df2['이동거리'], q=3)
````

````python
df2['이동거리_qcut'].value_counts()
````

구간도 예쁘게 분할(**균등하게 분할**)이 된 것 처럼 보입니다. 하지만, **간격은 일정하지 않다.**

qcut 또한 임의 범위를 조정할 수 있다.

````python
qcut_bins = [0, 0.2, 0.8, 1]
````

````python
pd.qcut(df2['이동거리'], qcut_bins)
````

qcut 역시 label을 지정할 수 있다. 마찬가지로 범위 수보다 1개 적게 설정한다.

````python
qcut_labels = ['적음', '보통', '많음']
````

````python
pd.qcut(df2['이동거리'], qcut_bins, labels=qcut_labels).value_counts()
````

나눈 다음 `age_bin` 컬럼을 생성하여 나눈 값을 대입하고 분포를 출력하시오

`cut()` 함수를 활용하시오

````python
# ================================================================== #
#         START OF YOUR CODE (DO NOT DELETE/MODIFY THIS LINE)        #
# ================================================================== #



# ================================================================== #
#          END OF YOUR CODE (DO NOT DELETE/MODIFY THIS LINE)         #
# ================================================================== #
````

````python
# ================================================================== #
#         START OF YOUR CODE (DO NOT DELETE/MODIFY THIS LINE)        #
# ================================================================== #

# age 컬럼의 최대값 확인 (결측치 제거 후)
max_age = sample["age"].max()

# 구간 설정 (0세 초과 ~ 15세 이하, 15세 초과 ~ 30세 이하, 30세 초과 ~ 45세 이하, 45세 초과 ~ 최대값)
bins = [0, 15, 30, 45, max_age]

# 각 구간의 이름 설정
labels = ["0~15", "15~30", "30~45", "45+"]

# cut() 함수를 사용하여 age 컬럼을 구간별로 나누기
sample["age_bin"] = pd.cut(sample["age"], bins=bins, labels=labels, right=True)

# age_bin 컬럼 분포 출력
print(sample["age_bin"].value_counts())

# ================================================================== #
#          END OF YOUR CODE (DO NOT DELETE/MODIFY THIS LINE)         #
# ================================================================== #
````

나이를 **3개의 균등 분할**을 갖도록 구간을 나누고 분포를 출력하시오
- 나이가 어린 그룹부터 차례대로 'young', 'normal', 'old' 라벨을 부여하시오

`qcut()` 함수를 활용하시오

````python
# ================================================================== #
#         START OF YOUR CODE (DO NOT DELETE/MODIFY THIS LINE)        #
# ================================================================== #



# ================================================================== #
#          END OF YOUR CODE (DO NOT DELETE/MODIFY THIS LINE)         #
# ================================================================== #
````

````python
# ================================================================== #
#         START OF YOUR CODE (DO NOT DELETE/MODIFY THIS LINE)        #
# ================================================================== #

# 결측치 제거 (qcut 사용 시 NaN이 있으면 오류 발생)
sample = sample.dropna(subset=["age"])

# 3개 구간으로 균등하게 나누고 라벨 지정
sample["age_group"] = pd.qcut(sample["age"], q=3, labels=["young", "normal", "old"])

# 각 그룹별 데이터 개수 출력
print(sample["age_group"].value_counts())

# ================================================================== #
#          END OF YOUR CODE (DO NOT DELETE/MODIFY THIS LINE)         #
# ================================================================== #
````
