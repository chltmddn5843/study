## 4. median() - 중앙값

데이터의 중앙 값을 출력 한다.

데이터를 **오름차순 정렬하여 중앙에 위치한 값**이다.

이상치(outlier)가 존재하는 경우, `mean()`보다 `median()`을 대표값으로 더 **선호**한다.

````python
pd.Series([1, 2, 3, 4, 5]).median()
````

````python
pd.Series([4, 5, 1, 2, 3]).median()
````

**짝수**개의 데이터가 있는 경우에는 **가운데 2개 중앙 데이터의 평균 값을 출력** 한다.

````python
pd.Series([1, 2, 3, 4, 5, 6]).median()
````

나이의 평균(mean)과 중앙값(median)은 약간의 **차이가 있음**을 확인할 수 있다.

````python
print(f"나이 평균: {df['age'].mean():.5f}\n나이 중앙값: {df['age'].median()}\n차이: {df['age'].mean() - df['age'].median():.5f}")
````

## 7. var() - 분산

$\large 분산 = \Huge\frac{\sum_{i=1}^{n}(X_{i} - \bar{X})^{2}}{n-1}$

$\large 평균 = \huge \bar{x}$

````python
# 평균
fare_mean = df['fare'].values.mean()
fare_mean
````

````python
# 분산
my_var = ((df['fare'].values - fare_mean) ** 2).sum() / (df['fare'].count() - 1)
my_var
````

다음과 같이 간단하게 분산을 구할 수 있다.

````python
df['fare'].var()
````

## 8. std() - 표준편차

$\large 표준편차 =\huge \sqrt{분산} = \sqrt{\frac{{}\sum_{i=1}^{n}(X_{i} - \bar{X})^{2}}{n-1}}$

분산(var)의 제곱근

````python
np.sqrt(df['fare'].var())
````

````python
df['fare'].std()
````

## 14. corr() - 상관관계

`corr()`로 컬럼(column)별 **상관관계**를 확인할 수 있다.

> **-1~1 사이의 범위**를 가진다.
>
> **-1에 가까울 수록 반비례** 관계, **1에 가까울수록 정비례** 관계를 의미한다.

corr함수는 숫자형 데이터에서만 작동하기 때문에, `numeric_only=True`를 붙여준다.

````python
df.corr(numeric_only=True)
````

**특정 컬럼에 대한 상관관계**를 확인할 수 있다.

````python
df.corr(numeric_only=True)['survived']
````
