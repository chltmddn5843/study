#1. 딥러닝 프레임워크를 왜 써야할까?

>**프레임워크(Framework)**란 소프트웨어 개발을 위해 특정한 목적을 위해 만들어진 구조체이다.
>
>**딥러닝 프레임워크(Deep Learning Framework)**는 딥러닝 모델 구현 및 학습을 위한 프레임워크로서, 다양한 딥러닝 모델 구현 및 학습 관련 함수를 제공한다.
>
>이를 통해 딥러닝 모델을 구현하여 활용하고자 하는 연구자와 개발자들은 중복된 작업이 줄어들어 핵심 작업에만 집중할 수 있다.


#2. 딥러닝 프레임워크 소개

>대표적인 딥러닝 프레임워크인 텐서플로우(TensorFlow)와 파이토치(PyTorch)를 소개한다.

##2-1. 텐서플로우(TensorFlow)

download.png

>TensorFlow는 구글이 개발한 오픈소스 프레임워크로 머신러닝과 딥러닝 관련 모델 구현 및 학습에 쉽게 사용할 수 있도록 여러 기능을 제공하고 있다.
>
>최근 입문자를 위한 기능을 강조한 keras 프레임워크와 병합되며, 다양한 수준의 사용자들이 텐서플로우 프레임워크를 이용할 수 있도록 돕고 있다.

##2-2. 파이토치(PyTorch)

download (2).png

>PyTorch는 페이스북(현 메타) 인공지능 연구팀에서 개발한 오픈소스 프레임워크이다. 
>
>직관적인 코드 작성이 가능해 현재 가장 많은 사람들이 이용하고 있는 딥러닝 프레임워크이다.

#3. PyTorch 딥러닝 프레임워크의 장점

##3-1. 사용성

>TensorFlow의 경우 코드 실행시 computation graph를 먼저 생성한 이후, 별도의 실행 세션을 만들어 생성된 graph에 input을 입력하는 **define-by-run** 방식을 채택했다.
>
>이러한 방식은 Python 기반 프로그래밍에 있어 직관적이지 않고, 디버깅이 어렵다는 단점이 있다. -> 파이썬 어려움
>
>PyTorch의 경우, 코드를 실행하면 별도의 실행 세션 없이 computation graph를 자동으로 생성한다.
>
>이러한 방식은 사용자들이 직관적으로 이해하기 쉽고, 코드에서 에러가 발생했을 경우 **디버깅이 쉽다**는 장점이 있다.
>
>Tensorflow는 define-by-run 방식의 단점을 개선하기 위해 TensorFlow 2.0 버전을 출시했으나 여전히 편의성 측면에서 PyTorch를 따라가지 못하고 있다는 평이 많다.

## 3-2. 커뮤니티와 지원

>PyTorch는 현재 가장 많은 인공지능 연구자와 개발자들이 사용하는 딥러닝 프레임워크이다.
>
>많은 사용자들이 관련 커뮤니티에서 활동하며 많은 레퍼런스를 공유하고 있고, stackoverflow 등의 커뮤니티에서는 PyTorch 공식 문서의 내용만으로는 파악하기 어려운 내용들을 사용자들끼리 주고 받으며 PyTorch을 더욱 사용하기 쉽게 만들고 있다.
>
> 실제로 Google trend에서 검색량 추이를 보면 2020년부터 PyTorch와 TensorFlow가 비슷한 수준이 되었고, 2021년 이후로는 PyTorch가 TensorFlow를 크게 앞선 것을 확인할 수 있다.

#4. PyTorch 시작하기


>일반적으로 딥러닝 모델 구현 및 활용을 위해 개발자 및 연구자들이 가장 많이 활용하는 사이트는 [Stack Overflow](https://stackoverflow.com/) 및 [PyTorch Documentation](https://pytorch.org/docs/stable/index.html)이다. 
>
>PyTorch의 공식 문서인 PyTorch Documentation의 경우 가장 정확한 문법과 사용법을 제공하고 있고 초급자를 위한 [Tutorial](https://pytorch.org/tutorials/)과 다양한 [예시](https://pytorch.org/tutorials/beginner/pytorch_with_examples.html)들을 제공하고 있어 기본적인 내용을 숙지하면 PyTorch 딥러닝 프레임워크에 빠르게 익숙해질 수 있다.
>
>해결하기 어려운 문제나 원인을 알 수 없는 에러가 발생한 경우 해당 문제나 출력된 문구 등으로 구글링하면 대부분 관련된 Stack Overflow 페이지를 찾을 수 있고, 또 직접 질문을 올려볼 수도 있다.
>
>검색을 이용해 문제 해결이 어려운 경우, [PyTorch Forum](https://discuss.pytorch.org/)에 질문을 올려 PyTorch 개발진들에게 직접 답변을 받아볼 수 있다.
